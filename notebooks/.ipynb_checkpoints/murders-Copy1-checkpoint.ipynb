### 2.5 Index creation

```python
# ****************************************************************************************************************************
# Création d'un index imbriqué. Il faut répéter les dates autant de fois qu'il y a de pays, et copier la liste des pays autant 
# de fois qu'il y a d'années
# See example : https://pandas.pydata.org/pandas-docs/stable/advanced.html
# ****************************************************************************************************************************

dates = [ list(itertools.repeat(year, len(gini_df.columns.values)))  for year in range(1978, 2006)]
dates = [item for sublist in dates for item in sublist] #Flatten the list
#continents = [ list(itertools.repeat(continent, len(countries_dico[continent]))) for continent in countries_dico.keys() ]

countries = list(gini_df.columns.values) * (len(gini_df.index.values))
#countries = [ country for continent in countries_dico.keys() for country in countries_dico[continent] ]
#print(countries)
arrays = [ dates, countries]

# Creates list of tuples like [ (1979, "Afghanistan"), (1979, "Albania"), (1979, "Algeria"), ..., (1980, "Afghanistan"), ...]
tuples = list(zip(*arrays)) 

index = pd.MultiIndex.from_tuples(tuples, names=['year', 'country'])
```

```python
#Concatene toutes les lignes des dataframes gini puis murder
gini_values = [item for sublist in gini_df.values for item in sublist] #Flatten the list
murder_values = [item for sublist in murder_df.values for item in sublist] #Flatten the list


def remove_comma(s):
    """
    Remplace les virgules d'une chaîne de caractères par des points.
    Utile pour les nombres décimaux écrits avec une virgule qui devront être castés de String vers Float
    """
    if not (type(s) is str or type(s) is np.str_):
        # just returns it untouched
        return s
    return s.replace(",", ".")

murder_values = list(map(remove_comma, murder_values))
```

```python
# **********************************************************************************
# Crée un dataframe contenant chaque ligne du df précédant avec un index imbriqué
# **********************************************************************************
df1 = pd.DataFrame(gini_values, index=index, columns =["Gini"])
df2 = pd.DataFrame(murder_values, index=index, columns = ["Murder"])
result = pd.concat([df1, df2], axis = 1)
```

```python
result = result.reset_index()
```

## 3. Plot 
### 3.1 Horizontal bar plot

```python
def plot_barh_by(continent, year):
    f = plt.figure(1, figsize=(15,10))

    y1 = result[ (result['year'] == year) & (result['Continent'] == continent) ]["Gini"]
    x = result[ (result['year'] == year) & (result['Continent'] == continent) ]["country"]

    plt.barh( range(len(y1)), y1.astype(float), align='center' )

    plt.yticks(range(len(y1)), x, size='small')
    plt.title("Gini index in {} in {}".format(continent, year))
    plt.show()
    
plot_barh_by("Europe", 2000)
```

### 3.2 Scatter plot

```python
def plot_gini_murder(df, continent, year):
    f = plt.figure(1, figsize=(8,5))

    #Selectionne les lignes correspondant à continent et year
    my_filter = [ (df['year'] == year) & (df['Continent'] == continent) ][0].values
    y1 = df[my_filter]["Gini"]
    y2 = df[my_filter]["Murder"]
    x = df[my_filter]["country"]
    
    #plt.plot( range(len(y1[y1mask])), y1[y1mask].astype(float), linestyle='-', marker='o', label="Gini" )
    #plt.plot( range(len(y2[y2mask])), y2[y2mask].astype(float) ,linestyle='-', marker='o', label="Murder")
    
    #Crée un masque qui capture les lignes qui ont un gini index ou murder rate null
    y1mask = [idx for idx, item in zip(y1.index , y1) if not math.isnan(item)]
    y2mask = [idx for idx, item in zip(y2.index , y2) if not math.isnan(float(item))]
    common_mask = set(y1mask).union(set(y2mask))
    
    plt.scatter( y2[common_mask], y1[common_mask].astype(float), marker='o' )

    #plt.yticks(range(len(y1)), x, size='small')
    plt.xlabel("Murder rate")
    plt.ylabel("Gini Index")
    plt.title("Gini index et murder rate en {} en {}".format(continent, year))
    plt.legend()
    plt.show()
    
plot_gini_murder(result, "Europe", 2000)
```
